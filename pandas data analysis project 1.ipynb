# finding the time when people gives more tips


```python
df.groupby(('day'))['tip'].mean()
```

```python
df.groupby(('time'))['tip'].mean()
```

# Plot these data


```python
df.groupby(('day'))['tip'].mean().plot.bar()
```

```python
df.groupby(('time'))['tip'].mean().plot.bar()
```

# Finding the average tips with reference to Day, size, Smoker/non smoker,sex columns

```python
df.groupby(('size'))['tip'].mean()
```

```python
df.groupby(('size'))['tip'].mean().plot.bar()
```

```python
df.groupby(('sex'))['tip'].mean()
```

```python
df.groupby(('sex'))['tip'].mean().plot.bar()
```

## Cross table

```python
df.groupby(['sex','time'])['tip'].mean()
```

```python
df.groupby(['sex','time'])['tip'].mean().plot.bar()
```

# Visualize the relation between the tip and Total bill Using seaborn

# Bifurication of Data 

```python
sns.relplot(
    x='total_bill',y='tip',data=df,col='sex',hue='time')
```

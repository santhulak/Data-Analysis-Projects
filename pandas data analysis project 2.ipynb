# credit card that expires in 2020

```python
df.columns
```

```python
df['CC Exp Date']
```

```python
#Split the date wih'/' as delimiter and count
def expiry():
    count=0
    for d in df['CC Exp Date']:
        if d.split('/')[1]=='20':
            count=count+1
    print(count)
```

```python
expiry()
```

## Alternate method using slicing

```python
len(df[df['CC Exp Date'].apply(lambda x:x[3:]=='20')])
```

# Display the Top email providers in the dataset

```python
df['Email']
```

```python
emaillist=[]
for email in df['Email']:
    emaillist.append(email.split('@')[1])
#print(emaillist)
```

```python
#Create a new column to store the email providers
df['email_provider']=emaillist
```

```python
df['email_provider'].value_counts()
```

```python
df['email_provider'].value_counts().head()
```

```python
#Altrenate method:

df['Email'].apply(lambda x:x.split('@')[1]).value_counts().head()
```
